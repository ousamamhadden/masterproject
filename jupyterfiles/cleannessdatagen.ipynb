```python
import numpy as np 
import pandas as pd 


df = pd.read_csv('airbnb-listings.csv', sep=";")
df.head()
```

```python
df2 = df[df["Country"] == 'United States']
```

```python
df2['Reviews per Month'].to_numpy()[101]
```

```python
def describeColummn(df, columnName):
    print(df2[columnName].describe())
    print(df2[columnName].value_counts())
    print(df2[columnName].isna().sum())
describeColummn(df2, 'Review Scores Cleanliness')
```

```python
describeColummn(df2, 'Reviews per Month')
```

```python
textColumns = ['Summary','Space','Experiences Offered','Neighborhood Overview','Notes', 'Transit', 'Access', 'Interaction', 'House Rules','Review Scores Cleanliness']
```

```python
def checkLength(a):
    print(a.map(lambda x: len(str(x))>= 1000).sum())
```

```python
for c in textColumns:
    checkLength(df2[c])
```

```python
df2.columns
```

```python
df3 = df2[textColumns]
```

```python
random_subset = df3.sample(frac=0.1, random_state=42)
```
